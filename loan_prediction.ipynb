```python
from sklearn.model_selection import  train_test_split
```

```python
input_data=data.drop(columns=['loan_status'])
output_data=data['loan_status']
```

```python
input_data
```

```python
output_data
```

```python
#splitting dataset into train and test
```

```python
x_train,x_test,y_train,y_test=train_test_split(input_data,output_data,test_size=0.2)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
## scaling the data using standard scalar
## to improve performance
## it is present in sklearn only
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
## scaling of input dataset
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.fit_transform(x_test)
```

```python
##creating model using logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train_scaled,y_train)
model.score(x_test_scaled,y_test)
```

```python
# Steps to Achieve Higher Accuracy:
# Data Preprocessing:

# Ensure the data is well-cleaned and normalized.
# Handle missing values effectively.
# Remove irrelevant or noisy features.
# Feature Engineering:

# Add new features or combine existing ones to capture better patterns.
# Use feature selection techniques to retain the most informative features.
# Hyperparameter Tuning:

# Use Grid Search or Randomized Search to optimize model parameters.
# Ensemble Models:

# Combine multiple models (e.g., VotingClassifier or StackingClassifier) to leverage their strengths.
# Cross-Validation:

# Use cross-validation to avoid overfitting and assess the model's performance.
# Try Deep Learning: If traditional models don't achieve the desired accuracy, try neural networks using TensorFlow or PyTorch.
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

# 1. Random Forest Classifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(x_train_scaled, y_train)
rf_score = rf_model.score(x_test_scaled, y_test)
print(f"Random Forest Accuracy: {rf_score * 100:.2f}%")

# 2. Gradient Boosting Classifier
gb_model = GradientBoostingClassifier(random_state=42)
gb_model.fit(x_train_scaled, y_train)
gb_score = gb_model.score(x_test_scaled, y_test)
print(f"Gradient Boosting Accuracy: {gb_score * 100:.2f}%")

# 3. Support Vector Machine (SVM)
svc_model = SVC(kernel='rbf', C=1, gamma='auto', random_state=42)
svc_model.fit(x_train_scaled, y_train)
svc_score = svc_model.score(x_test_scaled, y_test)
print(f"SVM Accuracy: {svc_score * 100:.2f}%")

# 4. K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(x_train_scaled, y_train)
knn_score = knn_model.score(x_test_scaled, y_test)
print(f"KNN Accuracy: {knn_score * 100:.2f}%")

# 5. Hyperparameter Tuning (Random Forest Example)
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10]
}
grid_search = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=5, scoring='accuracy')
grid_search.fit(x_train_scaled, y_train)
best_rf_model = grid_search.best_estimator_
best_rf_score = best_rf_model.score(x_test_scaled, y_test)
print(f"Tuned Random Forest Accuracy: {best_rf_score * 100:.2f}%")
```

```python
# 1. Random Forest Classifier
# What It Is: An ensemble learning method that builds multiple decision trees and combines their results (via majority voting for classification) to make predictions.
# Key Strengths:
# Handles missing data and categorical variables well.
# Reduces overfitting by averaging multiple trees.
# Applications: Widely used for tabular data, fraud detection, and healthcare predictions.
# 2. Gradient Boosting Classifier
# What It Is: An ensemble model that builds decision trees sequentially, where each tree corrects the errors of the previous one. Gradient descent is used to minimize the loss function.
# Why It Worked Well:
# Focuses on hard-to-predict samples by assigning higher weights to their errors.
# Captures complex relationships in data effectively.
# Key Strengths:
# Excellent for imbalanced datasets or when high precision is needed.
# Highly customizable with parameters like learning rate, tree depth, and boosting iterations.
# Applications: Often used in competitions (e.g., Kaggle) for problems like loan default prediction, customer churn, and click-through rate prediction.
# 3. Support Vector Machine (SVM)
# What It Is: A classification algorithm that finds the optimal hyperplane to separate classes in the feature space.
# Why It Might Be Slightly Lower:
# Performs well with clear margins but struggles with overlapping classes or noisy data.
# Key Strengths:
# Works well for high-dimensional datasets.
# Can handle non-linear relationships using kernels.
# Applications: Image recognition, text classification, and bioinformatics.
# 4. K-Nearest Neighbors (KNN)
# What It Is: A simple algorithm that classifies samples based on the majority class of their nearest neighbors in the feature space.
# Why It Performed Lower:
# Sensitive to irrelevant features and requires a well-scaled dataset.
# Slower for large datasets due to its lazy learning approach.
# Key Strengths:
# Easy to implement and interpretable.
# Effective for small datasets with simple patterns.
# Applications: Recommendation systems, anomaly detection, and pattern recognition.
# Why Gradient Boosting Performed Best
# Captures Complex Patterns: By building trees sequentially, it focuses on hard-to-learn patterns that simpler models may miss.
# Resistant to Overfitting: Gradient Boosting balances bias and variance effectively through careful optimization.
# Customizability: The ability to tune hyperparameters like learning rate and tree depth gives it an edge over other models.
```
